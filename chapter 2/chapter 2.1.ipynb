```python
before = id(Y)
Y = Y + X
id(Y) == before
```

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y # 如果写Z = X + Y 就会创建新的对象，地址不同
print('id(Z):', id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
# +=、-=、*=、/= 等是Pytorch原地操作符，相当于X[:] = X + Y
```
